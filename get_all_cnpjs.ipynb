```python
from bs4 import BeautifulSoup as bs
import requests as req
import os
```

```python
import re
```

```python
url = 'http://idg.receita.fazenda.gov.br/orientacao/tributaria/cadastros/cadastro-nacional-de-pessoas-juridicas-cnpj/dados-abertos-do-cnpj'
```

```python
session = req.Session()
headers = {
    "User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_9_5)AppleWebKit 537.36 (KHTML, like Gecko) Chrome",
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,/;q=0.8"}
ref = session.get(url, headers=headers)
soup = bs(ref.text, "html.parser")
print(soup) # abrindo e mostrando a página inicial (url)
```

```python
links = soup.findAll('a',{'href':re.compile('^http://idg.receita.fazenda.gov.br/orientacao/tributaria/cadastros/cadastro-nacional-de-pessoas-juridicas-cnpj/consultas/download/Socios')})
# encontra todos os links com o perfil dos arquivos que queremos na página
```

```python
est = dict()
```

```python
for estado in links:
    est[estado.string] = estado.get('href')
est
```
